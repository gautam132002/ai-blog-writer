```python
import gradio as gr
from aitextgen import aitextgen
```

```python
ai = aitextgen(model="EleutherAI/gpt-neo-125M",to_gpu=True)
```

```python
def ai_text(inp):
  generated_text = ai.generate_one(max_length = 1000, prompt = inp, no_repeat_ngram_size = 3) #repetition_penalty = 1.9)
  #print(type(generated_text))
  return generated_text
```

```python
output_text = gr.outputs.Textbox()
gr.Interface(ai_text,"textbox", output_text, title="AI BLOG WRITER.",
             description="AI based blog/article generating webapp supported by aitextgen. *** Instruction :: type topic in INP and and then SUBMIT.***").launch()
```
